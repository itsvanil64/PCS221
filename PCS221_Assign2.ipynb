```python
#ques 1-->
import random

rows = 20000
cols = 10000

with open("large_matrix.txt", "w") as f:
    for i in range(rows):
        row = [str(random.randint(0, 100)) for _ in range(cols)]
        f.write(",".join(row) + "\n")

print("Data file created successfully.")
```

```python
#ques 2-->
import numpy as np

matrix = np.random.randint(1, 10, (5, 5))

first_row = matrix[0]

for i in range(1, len(matrix)):
    matrix[i] = first_row + i   # assigning based on first row

print(matrix)
```

```python
#ques 3-->
import numpy as np
from collections import defaultdict

matrix = np.random.randint(1, 6, (4, 4))

groups = defaultdict(list)

for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        groups[matrix[i][j]].append((i, j))

print("Matrix:\n", matrix)
print("\nGrouped Elements:")
for key, value in groups.items():
    print(key, "->", value)
```

```python
#ques 4-->
import numpy as np

A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

addition = A + B
subtraction = A - B

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Addition:\n", addition)
print("Subtraction:\n", subtraction)
```

```python
#ques 5-->
import numpy as np

n = 4

matrix = np.random.choice(range(0, 100, 2), (n, n))  # only even numbers

print("Matrix:\n", matrix)
```

```python
#ques 6-->
# Tuple matrix
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

row_sums = []

for row in matrix:
    row_sums.append(sum(row))

print("Row-wise sums:", row_sums)
```
